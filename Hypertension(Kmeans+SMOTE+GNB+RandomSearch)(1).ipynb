```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scores = []
davies_scores = []

# Range of clusters to test
cluster_range = range(2, 8)  # Test with 2 to 5 clusters

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X_norm)

    # Calculate the Silhouette Score
    silhouette_avg = silhouette_score(X_norm, cluster_labels)

    # Calculate the Davies-Bouldin Score
    davies_bouldin_avg = davies_bouldin_score(X_norm, cluster_labels)

    scores.append(silhouette_avg)
    davies_scores.append(davies_bouldin_avg)

# Plot the results
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('Silhouette Score', color=color)
ax1.plot(cluster_range, scores, 'o-', color=color, label='Silhouette Score')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Second y-axis
color = 'tab:red'
ax2.set_ylabel('Davies-Bouldin Score', color=color)
ax2.plot(cluster_range, davies_scores, 'o-', color=color, label='Davies-Bouldin Score')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Silhouette and Davies-Bouldin Scores for Different Cluster Counts')
plt.show()
```

** without SMOTE**

```python
def make_classification(x_train, x_test, y_train, y_test):
    accuracy, f1, auc = [], [], []
    random_state = 42

    classifiers = [
        DecisionTreeClassifier(random_state=random_state),
        GaussianNB(),
        AdaBoostClassifier(DecisionTreeClassifier(random_state=random_state)),
        RandomForestClassifier(random_state=random_state),
        GradientBoostingClassifier(random_state=random_state),
        LogisticRegression(random_state=random_state),
        XGBClassifier(random_state=random_state),
        LGBMClassifier(random_state=random_state),
    ]

    for classifier in classifiers:
        clf = classifier
        clf.fit(x_train, y_train)

        # Predictions and probabilities
        y_preds = clf.predict(x_test)
        y_probs = clf.predict_proba(x_test)

        # Metrics
        accuracy.append(((accuracy_score(y_test, y_preds)) * 100))
        f1.append(((f1_score(y_test, y_preds, average='weighted')) * 100))

        # Corrected ROC AUC with proper multi_class and average handling
        roc_auc = roc_auc_score(y_test, y_probs, multi_class='ovr', average='weighted')
        auc.append(roc_auc * 100)

    results_df = pd.DataFrame({
        "Accuracy Score": accuracy,
        "F1 Score": f1,
        "ROC Score": auc,
        "ML Models": ["DecisionTree","GNB" ,"AdaBoost", "RandomForest", "GradientBoosting", "LogisticRegression", "XGBoost", "LGBM"],
    })

    results = (results_df.sort_values(by=['ROC Score', 'F1 Score'], ascending=False).reset_index(drop=True))
    return classifiers, results

# Test with your data
classifiers, results = make_classification(x_train, x_test, y_train, y_test)

results
```

**ORG MODEL**

```python
models_org={
    "GNB": GaussianNB(),

}
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from imblearn.metrics import specificity_score

def test_model_performance_org(models_org, x_train, x_test, y_train, y_test):
    # Set the average parameter to avoid ValueError with multi-class data
    average_setting = 'macro'  # 'weighted' or 'micro' also valid

    for model_name, model in models_org.items():
        # Fit the model
        model.fit(x_train, y_train)

        # Get predictions and probabilities
        predicted = model.predict(x_test)
        predicted_proba = model.predict_proba(x_test)  # Ensure this gives probabilities

        # Calculate metrics with proper averaging for multi-class
        accuracy = accuracy_score(y_test, predicted) * 100
        sensitivity = recall_score(y_test, predicted, average=average_setting) * 100
        precision = precision_score(y_test, predicted, average=average_setting) * 100
        roc = roc_auc_score(y_test, predicted_proba, multi_class='ovr', average=average_setting) * 100
        f1score = f1_score(y_test, predicted, average=average_setting) * 100

        # Specificity calculation (check your implementation to ensure compatibility with multi-class)
        # Make sure this supports multi-class and uses correct `average_setting`
        specificity = specificity_score(y_test, predicted, average=average_setting) * 100  # Verify correctness

        # Output results
        print(f'{model_name}: Accuracy - {accuracy:.2f}%, Recall - {sensitivity:.2f}%, Precision - {precision:.2f}%, AUROC - {roc:.2f}%, F1_Score - {f1score:.2f}%, Specificity - {specificity:.2f}%')
```

```python
test_model_performance_org(models_org, x_train, x_test, y_train, y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, roc_auc_score
import numpy as np

# Create a plot with a fixed size
plt.figure(figsize=(10, 8))

# Determine the number of classes in your test dataset
num_classes = len(np.unique(y_test))

for name, clf in models.items():
    clf.fit(X_smote, y_smote)

    # Initialize lists for ROC data
    fpr_dict = {}
    tpr_dict = {}
    roc_auc_dict = {}

    for i in range(num_classes):
        # If multi-class, get the decision function or probability for each class
        if hasattr(clf, "decision_function"):
            y_score = clf.decision_function(x_test)
            y_score_class = y_score[:, i]
        else:
            y_score = clf.predict_proba(x_test)
            y_score_class = y_score[:, i]

        # Generate ROC curve for each class
        fpr_dict[i], tpr_dict[i], _ = roc_curve(y_test == i, y_score_class)
        roc_auc_dict[i] = auc(fpr_dict[i], tpr_dict[i])

        # Plot the ROC curve
        plt.plot(fpr_dict[i], tpr_dict[i], label=f'{name} - Class {i} (AUC = {roc_auc_dict[i]:.4f})')

# Add a diagonal line for reference
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

# Set plot limits
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

# Add labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for Multiple Classes')

# Show legend and display the plot
plt.legend(loc="lower right")
plt.show()
```
